```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
%matplotlib inline
```

```python
dataset = datasets.load_iris()
feature_names = dataset.feature_names
X = dataset.data
y = dataset.target
```

```python
fig = plt.figure(1, figsize=(16,9))
ax = fig.add_subplot(111, projection='3d',azim=120)

decomp = decomposition.PCA(n_components=3)
decomp.fit(X)
X = decomp.transform(X)

for name, label in [('Setosa',0),('Versicolor',1),('Virginica',2)]:
  ax.text3D(
      X[y == label, 0].mean(),
      X[y == label, 1].mean() + 1.5,
      X[y == label, 2].mean(),
      name,
      horizontalalignment="center",
      bbox=dict(alpha=0.5, edgecolor="w", facecolor="w"),
  )
y = np.choose(y, [1, 2, 0]).astype(float)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='magma', edgecolor="k")

plt.show()
```
